```python
df.drop('id', axis = 1, inplace = True)
df.columns = ['age', 'blood_pressure', 'specific_gravity', 'albumin', 'sugar', 'red_blood_cells', 'pus_cell',
              'pus_cell_clumps', 'bacteria', 'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium',
              'potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count',
              'hypertension', 'diabetes_mellitus', 'coronary_artery_disease', 'appetite', 'peda_edema',
              'anemia', 'class']
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
df["packed_cell_volume"] = pd.to_numeric(df["packed_cell_volume"],errors = "coerce")
df["white_blood_cell_count"] = pd.to_numeric(df["white_blood_cell_count"],errors = "coerce")
df["red_blood_cell_count"] = pd.to_numeric(df["red_blood_cell_count"],errors = "coerce")
```

```python
cat_col = [col for col in df.columns if df[col].dtype == "object"]
num_col = [col for col in df.columns if df[col].dtype != "object"]
```

```python
def solve_my_random_value(feature):
    random_sample = df[feature].dropna().sample(df[feature].isna().sum())
    random_sample.index = df[df[feature].isnull()].index
    df.loc[df[feature].isnull(),feature] = random_sample

for col in num_col:
    solve_my_random_value(col)

def solve_my_mode(feature):
    mode = df[feature].mode()[0]
    df[feature] = df[feature].fillna(mode)

for col in cat_col:
    solve_my_mode(col)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
encoders = {}

for cols in cat_col:
  le = LabelEncoder()
  df[cols] = le.fit_transform(df[cols])
  encoders[cols] = le
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
correlation_matrix = df.corr()

target_col = "class"

threshold = 0.25

high_corr_features = correlation_matrix[target_col][correlation_matrix[target_col].abs()>threshold].index.tolist()

high_corr_features.remove(target_col)   

high_corr_features
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler

scaler = {}

for col in high_corr_features:
    ssc = StandardScaler()
    df[col] = ssc.fit_transform(df[[col]])
    scaler[col] = ssc

scaler
```

```python
x,y = df[high_corr_features],df["class"]

y.value_counts()
```

```python
# Balancing of Data

from imblearn.over_sampling import RandomOverSampler

over_sampling = RandomOverSampler()

x_resample , y_resample = over_sampling.fit_resample(x,y)

# from imblearn.under_sampling import RandomUnderSampler

# under_sampling = RandomUnderSampler()

# x_resample , y_resample = under_sampling.fit_resample(x,y)
```

```python
y_resample.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(x_resample,y_resample,random_state=42)
```

```python
from sklearn.metrics import accuracy_score,classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

models = {"LogisticRegression" : LogisticRegression(),
          "DecisionTreeClassifier" : DecisionTreeClassifier(),
          "RandomForestClassifier" : RandomForestClassifier(),
          "GradientBoostingClassifier" : GradientBoostingClassifier(),
          "SVC" : SVC(),
          "KNeighborsClassifier" : KNeighborsClassifier()}

results = {}

for name, model in models.items():
    model.fit(X_train,Y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(Y_test,y_pred)
    results[name] = acc
    print(f"\n{name} Accuracy: {acc:.4f}")
    print(classification_report(Y_test, y_pred))

best_model = max(results, key=results.get)
print("\n Best Model is:", best_model, "with Accuracy:", results[best_model])
```

```python
final_model = models[best_model]

final_model
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test,final_model.predict(X_test))

cm
```
