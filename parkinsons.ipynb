```python
correlation_matrix = df.corr()

target_col = "status"

threshold = 0.3

high_corr_features = correlation_matrix[target_col][correlation_matrix[target_col].abs()>threshold].index.tolist()

high_corr_features.remove(target_col)   

high_corr_features
```

```python
data = df.copy()

data.head()
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler

scaler = {}

for col in high_corr_features:
    ssc = StandardScaler()
    data[col] = ssc.fit_transform(data[[col]])
    scaler[col] = ssc

scaler

```

```python
X,Y = data[high_corr_features],data["status"]

Y.value_counts()
```

```python
# Balancing of Data

from imblearn.over_sampling import RandomOverSampler

over_sampling = RandomOverSampler()

X_resample , Y_resample = over_sampling.fit_resample(X,Y)

# from imblearn.under_sampling import RandomUnderSampler

# under_sampling = RandomUnderSampler()

# x_resample , y_resample = under_sampling.fit_resample(x,y)
```

```python
Y_resample.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X_resample,Y_resample,random_state=42)
```

```python
print(f"x_train shape: {X_train.shape}")
print(f"x_test shape: {X_test.shape}")
print(f"y_train shape: {Y_train.shape}")
print(f"y_test shape: {Y_test.shape}")
```

```python
from sklearn.metrics import accuracy_score,classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

models = {"LogisticRegression" : LogisticRegression(),
          "DecisionTreeClassifier" : DecisionTreeClassifier(),
          "RandomForestClassifier" : RandomForestClassifier(),
          "GradientBoostingClassifier" : GradientBoostingClassifier(),
          "SVC" : SVC(),
          "KNeighborsClassifier" : KNeighborsClassifier()}

results = {}

for name, model in models.items():
    model.fit(X_train,Y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(Y_test,y_pred)
    results[name] = acc
    print(f"\n{name} Accuracy: {acc:.4f}")
    print(classification_report(Y_test, y_pred))

best_model = max(results, key=results.get)
print("\n Best Model is:", best_model, "with Accuracy:", results[best_model])
```

```python
final_model = models[best_model]

final_model
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test,final_model.predict(X_test))

cm
```
